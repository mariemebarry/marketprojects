```python
# import pandas package; use pandas read_excel() to read the .xlsx data into a pandas DataFrame

import pandas as pd
df = pd.read_excel('GOOG.xlsx')
df
```

```python
# Using the pandas DataFrame created in previous set, calculate and append a column for daily returns. Use AdjClose 
# prices when calculating daily returns. The pandas method pct_change() can be used to calculate daily returns.

df['Returns'] = ((df[['Open', 'Adj Close']].pct_change(axis=1)['Adj Close']) * 100).round(2).map(str) + '%'
df
```

```python
# create a histogram for the daily returns calculated in step A.3, using pandas or matplotlib

df['Returns'] = df['Returns'].str.rstrip('%').astype('float') / 100.0
df.Returns.plot(kind='hist')
```
