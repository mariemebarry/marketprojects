```python
# Part A
# write Python statements to query the tickers table in the dj.db database

import sqlite3 as db
import pandas as pd

connection = db.connect('dj.db')

cursor = connection.cursor()

cursor.execute('SELECT id, ticker, name FROM tickers')

# For this part, use fetchall(). Place the results in a pandas DataFrame

rows = cursor.fetchall()

df = pd.DataFrame(rows)

df.columns = ['id', 'ticker', 'company']

# use the ticker field as theDataFrame row index
df.set_index('ticker', inplace=True)

# display the DataFrame
df
```

```python
# Part B
# Write Python statements to query the daily_measures table in the dj.db database,
# to determine the earliest and latest date of the data in this table

# earliest date: 01-03-2017
# latest date: 12-29-2017

import sqlite3 as db

connection = db.connect('dj.db')

cursor = connection.cursor()

cursor.execute('SELECT min(date) as first, max(date) as last FROM daily_measures')

for row in cursor:
   print(row)
```

```python
# Part D
# write Python statements to determine the date and price when Apple had the highest one-day daily
# percent change. Use the same SQL statement as in part C, and additional Python (and possibly pandas) 
# statements to find the result


import sqlite3 as db
import pandas as pd

connection = db.connect('dj.db')

cursor = connection.cursor()

cursor.execute('SELECT date, closing FROM daily_measures WHERE ticker_id=3 ORDER BY date')

rows = cursor.fetchall()

df = pd.DataFrame(rows)

df.columns = ['date', 'closing']

df.set_index('date', inplace=True)

df['DailyPercentChange'] = ((df['closing'] - df['closing'].shift(1)) / df['closing'].shift(1) * 100)

df
```

```python
# display the date with the highest daily percent change for AAPL

# date: 02-01-2017
# daily percent change value: 6.098063%

df.loc[df['DailyPercentChange'].idxmax()]
```
